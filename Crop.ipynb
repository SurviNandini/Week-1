```python
# Extract feature column names from the dataset and remove the 'label' column from the list
features = crop.columns.to_list()
features.remove('label')
print (features)
```

```python
# Visualize the distribution of each feature
import seaborn as sns
fig, ax = plt.subplots(3, 3, figsize = (15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(crop[i], ax=subplot, kde=True)
    subplot.set_title(i)
plt.tight_layout()
plt.show()
```

```python
# Plotting scatter plot
import seaborn as sns
fig, ax = plt.subplots (3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.scatterplot(x=i, y='label', data=crop, ax=subplot)
plt.tight_layout()
plt.show()
```

```python
# Finding outliers boxplot
fig, ax = plt.subplots(3,3, figsize = (15,15))
for i, subplot in zip(features, ax.flatten()):
    sns.boxplot(crop[i], ax=subplot)
plt.tight_layout()
plt.show()
```

```python
# Correlation of the features
num_cols=crop.select_dtypes(include = [np.number])
num_cols
```

```python
# Compute the correlation matrix for numerical columns
corr = num_cols.corr()
corr
```

```python
# Visualizing with Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Encoding
crop['label'].unique()
```

```python
# Dictionary mapping crop names to unique integer labels
crop_dict = {
    'rice': 0,
    'maize': 1,
    'chickpea': 2,
    'kidneybeans': 3,
    'mungbean': 4,
    'blackgram': 5,
    'lentil': 6,
    'pomegranate': 7,
    'banana': 8,
    'mango': 9,
    'grapes': 10,
    'watermelon': 11,
    'muskmelon': 12,
    'apple': 13,
    'orange': 14,
    'papaya': 15,
    'coconut': 16,
    'cotton': 17,
    'jute': 18,
    'coffee': 19
}
```

```python
# Map crop labels to their corresponding numerical values using crop_dict
crop['crop_no'] = crop['label'].map(crop_dict)
crop.head()
```

```python
crop.tail()
```

```python
# Drop the 'label' column from the dataset as it is now mapped to numerical values
crop.drop('label', axis=1, inplace=True)
crop.tail()
```
